**Statistics Advance Part 1**

1. **What is a random variable in probability theory?**
   - A random variable is a numerical outcome of a random experiment. It assigns a number to each possible outcome in a sample space.

2. **What are the types of random variables?**
   - **Discrete Random Variable:** Takes a finite or countable number of values (e.g., number of heads in coin flips).
   - **Continuous Random Variable:** Takes an infinite number of values within a range (e.g., height, weight).

3. **What is the difference between discrete and continuous distributions?**
   - Discrete distributions have a countable number of outcomes (e.g., Binomial, Poisson), while continuous distributions have an infinite number of outcomes in a given range (e.g., Normal, Uniform).

4. **What are probability distribution functions (PDF)?**
   - A probability distribution function describes the likelihood of each possible value a random variable can take.

5. **How do cumulative distribution functions (CDF) differ from probability distribution functions (PDF)?**
   - PDF gives the probability of a specific outcome, while CDF gives the cumulative probability up to a certain value.

6. **What is a discrete uniform distribution?**
   - A discrete uniform distribution is where all outcomes have equal probability (e.g., rolling a fair die).

7. **What are the key properties of a Bernoulli distribution?**
   - It has two possible outcomes: success (1) and failure (0) with probability p and (1-p) respectively.

8. **What is the binomial distribution, and how is it used in probability?**
   - It models the number of successes in a fixed number of independent Bernoulli trials.

9. **What is the Poisson distribution and where is it applied?**
   - It models the number of events in a fixed interval when events occur independently (e.g., calls received in an hour).

10. **What is a continuous uniform distribution?**
    - A distribution where all values in a given interval are equally likely.

11. **What are the characteristics of a normal distribution?**
    - Bell-shaped, symmetric, mean = median = mode, defined by mean (μ) and standard deviation (σ).

12. **What is the standard normal distribution, and why is it important?**
    - It is a normal distribution with mean 0 and standard deviation 1, used for Z-score calculations.

13. **What is the Central Limit Theorem (CLT), and why is it critical in statistics?**
    - CLT states that the distribution of the sample mean approaches a normal distribution as the sample size increases, regardless of population distribution.

14. **How does the Central Limit Theorem relate to the normal distribution?**
    - It explains why normal distribution is commonly observed in statistical data analysis.

15. **What is the application of Z statistics in hypothesis testing?**
    - Z-statistics is used to determine if a sample mean significantly differs from a population mean.

16. **How do you calculate a Z-score, and what does it represent?**
    - Z = (X - μ) / σ. It represents how many standard deviations a value is from the mean.

17. **What are point estimates and interval estimates in statistics?**
    - Point estimate: A single value estimate of a parameter.
    - Interval estimate: A range of values (confidence interval) within which the parameter is likely to lie.

18. **What is the significance of confidence intervals in statistical analysis?**
    - It provides a range of plausible values for an unknown parameter with a specified confidence level.

19. **What is the relationship between a Z-score and a confidence interval?**
    - A Z-score helps determine the confidence interval width and margin of error.

20. **How are Z-scores used to compare different distributions?**
    - Z-scores standardize data, allowing comparison across different distributions.

21. **What are the assumptions for applying the Central Limit Theorem?**
    - Random sampling, independent observations, sufficiently large sample size (usually n > 30).

22. **What is the concept of expected value in a probability distribution?**
    - It is the long-run average value of a random variable (E(X) = Σ[x * P(x)]).

23. **How does a probability distribution relate to the expected outcome of a random variable?**
    - The probability distribution defines the likelihood of each outcome, determining the expected value.



```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as stats
```

```python
# 1. Generate a random variable and display its value
random_variable = np.random.randint(1, 101)
print(f"Random Variable: {random_variable}")
```

```python
# 2. Generate a discrete uniform distribution and plot PMF
x = np.arange(1, 7)
p = [1/6] * 6
plt.bar(x, p)
plt.title("Discrete Uniform Distribution (Die Roll)")
plt.xlabel("Outcome")
plt.ylabel("Probability")
plt.show()
```

```python
# 3. Bernoulli distribution PDF
p = 0.5
def bernoulli_pdf(x, p):
    return p**x * (1-p)**(1-x)

x_values = [0, 1]
y_values = [bernoulli_pdf(x, p) for x in x_values]
plt.bar(x_values, y_values)
plt.title("Bernoulli Distribution PDF")
plt.show()
```

```python
# 4. Binomial distribution simulation
n, p = 10, 0.5
binomial_data = np.random.binomial(n, p, 1000)
plt.hist(binomial_data, bins=np.arange(n+2)-0.5, density=True, alpha=0.75)
plt.title("Binomial Distribution Histogram (n=10, p=0.5)")
plt.show()
```

```python
# 5. Poisson distribution visualization
poisson_data = np.random.poisson(3, 1000)
plt.hist(poisson_data, bins=range(10), density=True, alpha=0.75)
plt.title("Poisson Distribution")
plt.show()
```

```python

# 6. Continuous uniform distribution visualization
uniform_data = np.random.uniform(0, 1, 1000)
plt.hist(uniform_data, bins=20, density=True)
plt.title("Continuous Uniform Distribution")
plt.show()
```

```python
# 7. Normal distribution simulation
normal_data = np.random.normal(0, 1, 1000)
plt.hist(normal_data, bins=30, density=True)
plt.title("Normal Distribution Histogram")
plt.show()
```

```python
# 8. Z-score calculation and visualization
def z_score(data):
    return (data - np.mean(data)) / np.std(data)

z_scores = z_score(normal_data)
plt.hist(z_scores, bins=30, density=True)
plt.title("Z-Score Distribution")
plt.show()
```

```python
# 9. Central Limit Theorem demonstration
sample_means = [np.mean(np.random.exponential(1, 30)) for _ in range(1000)]
plt.hist(sample_means, bins=30, density=True)
plt.title("Central Limit Theorem Demonstration")
plt.show()
```

```python

# 16. Standard normal distribution plot
x = np.linspace(-4, 4, 100)
y = stats.norm.pdf(x, 0, 1)
plt.plot(x, y)
plt.title("Standard Normal Distribution (Mean=0, Std=1)")
plt.show()
```

```python
# 17. Generate binomial random variables and compute probabilities
binomial_values = np.random.binomial(n=10, p=0.5, size=1000)
probabilities = stats.binom.pmf(binomial_values, n=10, p=0.5)
plt.bar(binomial_values, probabilities)
plt.title("Binomial Distribution PMF")
plt.show()
```

```python

# 18. Compute and compare Z-score with standard normal distribution
data_point = 8
z = (data_point - np.mean(binomial_values)) / np.std(binomial_values)
print(f"Z-score of {data_point}: {z}")
```

```python
# 19. Hypothesis testing using Z-statistics
data = np.random.normal(50, 10, 30)
z_statistic, p_value = stats.ttest_1samp(data, 50)
print(f"Z-Statistic: {z_statistic}, P-Value: {p_value}")
```

```python
# 20. Confidence interval calculation
mean, std = np.mean(data), np.std(data, ddof=1)
confidence_interval = stats.norm.interval(0.95, loc=mean, scale=std/np.sqrt(len(data)))
print(f"95% Confidence Interval: {confidence_interval}")
```

```python
# 21. Generate and interpret confidence interval from normal distribution
sample = np.random.normal(100, 15, 50)
mean_sample, std_sample = np.mean(sample), np.std(sample, ddof=1)
ci_sample = stats.norm.interval(0.95, loc=mean_sample, scale=std_sample/np.sqrt(len(sample)))
print(f"Confidence Interval of Sample Mean: {ci_sample}")
```

```python

# 22. Probability Density Function (PDF) of a normal distribution
x = np.linspace(50, 150, 1000)
pdf = stats.norm.pdf(x, 100, 15)
plt.plot(x, pdf)
plt.title("Normal Distribution PDF")
plt.show()
```

```python

# 23. Cumulative Distribution Function (CDF) of Poisson distribution
poisson_cdf = stats.poisson.cdf(np.arange(0, 10), mu=3)
plt.plot(np.arange(0, 10), poisson_cdf, marker='o')
plt.title("Poisson Distribution CDF")
plt.show()
```

```python
# 24. Simulating continuous uniform distribution and expected value
cont_uniform = np.random.uniform(10, 20, 1000)
expected_value = np.mean(cont_uniform)
print(f"Expected Value of Continuous Uniform Distribution: {expected_value}")

```

```python
# 25. Compare standard deviations of two datasets
dataset1 = np.random.normal(50, 10, 1000)
dataset2 = np.random.normal(60, 15, 1000)
plt.hist(dataset1, bins=30, alpha=0.5, label="Dataset 1")
plt.hist(dataset2, bins=30, alpha=0.5, label="Dataset 2")
plt.legend()
plt.title("Comparison of Standard Deviations")
plt.show()
```

```python
# 26. Calculate range and interquartile range (IQR) of a dataset
data_normal = np.random.normal(100, 20, 1000)
data_range = np.ptp(data_normal)
iqr = stats.iqr(data_normal)
print(f"Range: {data_range}, IQR: {iqr}")
```

```python
# 27. Implement Z-score normalization and visualize transformation
z_normalized = (data_normal - np.mean(data_normal)) / np.std(data_normal)
plt.hist(z_normalized, bins=30, density=True)
plt.title("Z-score Normalized Data")
plt.show()
```

```python
# 28. Calculate skewness and kurtosis of a normal distribution dataset
skewness = stats.skew(data_normal)
kurtosis = stats.kurtosis(data_normal)
print(f"Skewness: {skewness}, Kurtosis: {kurtosis}")
```
